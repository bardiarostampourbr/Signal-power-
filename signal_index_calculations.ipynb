```python
# تعریف متغیرها
n = 5  # مقدار نمونه برای n
t = 2  # مقدار نمونه برای t
```

```python
# محاسبه x[n-2]
result_1 = n - 2
print(f"x[{n}-2] = {result_1}")
```

```python
# محاسبه x[3/2 n + 3]
result_2 = (3/2) * n + 3
print(f"x[{3/2*n}+3] = {result_2}")
```

```python
# محاسبه x[n] + x[2n] - 3
x = lambda i: i * 2  # تابع نمونه برای x[i]
result_3 = x(n) + x(2*n) - 3
print(f"x[{n}] + x[{2*n}] - 3 = {result_3}")
```

```python
# محاسبه x[3t - 1/2]
result_4 = 3 * t - 0.5
print(f"x[{3*t}-0.5] = {result_4}")
```

```python
# محاسبه x[1/2 t + 3]
result_5 = (1/2) * t + 3
print(f"x[{0.5*t}+3] = {result_5}")
```

```python
# محاسبه x[(t-2) - t - t]
result_6 = (t - 2) - t - t
print(f"x[{(t-2)}-t-t] = {result_6}")
```
